## Trigonometric function

A trigonometric function is displayed as follows:

trigonometric_0_1.png

As shown in the figure above, when the positive direction of the $x$ axis is taken as the starting line on the coordinate plane,<br>
if P(x,y) is the intersection of a circle with a radius of length $r$<br>
and a radius representing the general angle $θ$,<br>
and the origin O as the center,<br>
the values of $\cfrac{y}{r}$, $\cfrac{x}{r}$, $\cfrac{y}{x}$$(x\neq0)$ are set to one each according to the value of $θ$.<br><br>

Thus, the correspondences such as $θ\rightarrow\cfrac{y}{r}$, $θ\rightarrow\cfrac{x}{r}$, $θ\rightarrow\cfrac{y}{x}$$(x\neq0)$ are functions of $θ$.<br><br>

Now, let's call the functions of $θ$ as sine function, cosine function, and tangent function respectively.<br>
These functions are displayed in $sinθ\rightarrow\cfrac{y}{r}$, $cosθ\rightarrow\cfrac{x}{r}$, $tanθ\rightarrow\cfrac{y}{x}$$(x\neq0)$.<br><br>

We will define a class for the graph of a trigonometric function.<br>
Before we define the class, note that the method for trigonometric functions in python receives the radian value as its input.<br>
Thus, we need to define a function to convert the degree value to radian value.<br><br>

The equations to convert degree to radian and radian to degree are as follows:<br>
1. $Radians = [\cfrac{π}{180}]\times x degrees$<br>
2. $Degrees = [\cfrac{180}{π}]\times x raddians$<br><br>

$π$ can be implemented as follows:<br>
```python3
import math
print(math.pi)
# The output is 3.141592653589793
```

What we are going to do are:
1. Define the function 'degree_to_radian' to convert degree to radian.<br>
2. Complete the function 'TrigonometricEquation'.<br>
3. Using the function 'TrigonometricEquation' display graphs of
sine function, cosine function, and tangent function
within the value of dist is 5, then print them.<br>

The expected results are:

trigonometric_1.png

````python
import math
from matplotlib import pyplot as plt


class TrigonometricEquation:
    def __init__(self, trig_type, dist=1, size=360):
        self.size = size
        self.xs = []
        self.ys = []
        trig = None

        if trig_type == "sin":
            trig = math.sin
        if trig_type == "cos":
            trig = math.cos
        if trig_type == "tan":
            trig = math.tan

        x = -size

        while True:
            def degrees_to_radians(x):
                return (math.pi/180)*x

            self.xs.append(x)
            self.ys.append(trig(degrees_to_radians(x)))
            x += dist

            if x > size:
                break


    def show_graph(self):
        plt.grid(True)
        axes = plt.gca()
        axes.set_xlim([-360, 360])
        axes.set_ylim([-1, 1])
        plt.plot(self.xs, self.ys, '.')
        # plt.cla()  # This method initiates the graphs.


TrigonometricEquation(trig_type="sin", dist=5, size=360).show_graph()
````

````python
TrigonometricEquation(trig_type="cos", dist=5, size=360).show_graph()
````

````python
TrigonometricEquation(trig_type="tan", dist=5, size=360).show_graph()
````
