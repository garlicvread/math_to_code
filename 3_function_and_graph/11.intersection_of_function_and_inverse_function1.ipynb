## The intersection of a function and its inverse function (1)

The intersection of a function and its inverse is equal to the intersection of the original function and `y=x`.

Find the intersection of the function $ f(x) = 3x - 4 $ and its inverse function then return it.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (3 * x) - 4

# Inverse function: f^-1(x)=(x+4)/3
def f_inv(x):
    return (x + 4) / 3

# Identity function I(x): y=x
def identity(x):
    return(x)

# f=f^-1
x = np.linspace(-1, 5, 601)
graph, = plt.plot(x, f(x))
graph1, = plt.plot(x, f_inv(x))

graph2, = plt.plot(x, identity(x), '--')  # --: dash graph option
plt.legend(handles=(graph,graph1,graph2),  # Add a legend for the graph
           labels=(r'$ y=f(x) $', r'$ y=f^{-1}(x) $', r'$ y=x $'))

# Limit the range of y-axis.
plt.ylim(-1, 5)
plt.show()
```
