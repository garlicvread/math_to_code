## Logarithm function
The logarithm function, or the logarithmic function, or log function is a mathematical function denoted by $y=log_a(x)$.

In mathematics, the logarithm is the inverse function ot exponentiation. This means that the logarithm of a given number $x$ is the exponent to which another fixed number, the base $b$, must be raised, to produce that number $x$.

A log function $log_a(x)$ becomes as one value thus it is a function of $x$ when
- $a$ and $x$ are real numbers
- $a>0$ and $a\neq1$

In this case, we call this function as the logarithm of $a$ base.

Let's define a logarithm function which is displayed as $y=log_a(x+dx)+c$ as follows.


Within this class, the parameters $a$, $dx$ and $c$ are the same with $a$, $dx$ and $c$ of the logarithm function.

The logarithm function can be written with a while loop.

if we know $x$ coordinate of the point, then we can calculate the y coordinate of the point with values of $a$, $dx$, and $c$.

1. Set the $x$ coordinate of the point as the smallest value of the range of $x$ of the graph that will be displayed.

2. Then with the while loop, increase the value of $x$ as much as the value of $dist$, then store $y$ value that is calculated.

3. When the values of $x$ exceeds the range of the graph, then stop the loop.

NOTE: you need to use log method of math module to calculate the logarithm function.

e.g.) To calculate $log_28$:

```python3
import math
print(math.log(8, 2))
# The output will be 3.0.
```

````python
import math
from matplotlib import pyplot as plt


class LogarithmeticEquation:
    def __init__(self, a, dx=0, c=0, dist=1, size=100):
        self.size = size
        self.xs = []
        self.ys = []

        x = 0.001 # The domain of x is greater than 0. Set x as 0.001 to avoid 0.

        # # Variation 1:
        # while x < size:
        #     x = x + dist
        #     y = math.log(x+dx, a) + c
        #     self.xs.append(x)
        #     self.ys.append(y)

        # Variation 2:
        while True:
            self.xs.append(x)
            self.ys.append(math.log(x+dx, a) + c)
            x += dist

            if x > size:
                break


    def show_graph(self):
        plt.grid(True)
        axes = plt.gca()
        axes.set_xlim([-self.size, self.size])
        axes.set_ylim([-self.size, self.size])
        plt.plot(self.xs, self.ys, '.')


LogarithmeticEquation(a=2, dx=0, c=0, dist=0.1, size=10).show_graph()
````
