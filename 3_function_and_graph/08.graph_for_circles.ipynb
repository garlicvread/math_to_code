Graph for circles

Let's declare a class to draw a circle.

With this class, you can draw circles in the way of calculating the coordinates of the unit circle.
The radius of the unit circle is 1, thus you can calculate the coordinates of the circle using the values following.
$$(cosθ, sinθ)$$

The coordinate above is for the circle that has the center at the origin, and the radius is 1.
Thus, when you want to draw other circles with different centers, you can modify the above coordinates as follows.

$$((r * cosθ) + x, (r * sinθ) + y)$$

You can declare a class to draw a circle with the sin method and cos method of the math module that was used to calculate trigonometry.

For example, this is the code to calculate $cos0$.

```python3
import math

def degrees_to_radians(degrees):
    return degrees * (math.pi/180.0)

radians = degrees_to_radians(0)

print(math.cos(radians))  # The expected result is 1.0.
```

````python
#※들여쓰기에 주의하세요!
import math
from matplotlib import pyplot as plt


def degrees_to_radians(degrees):
    return degrees * (math.pi / 180.0)


class CircleEquation:
    def __init__(self, x=0, y=0, r=1, dist=1):  
        self.x = x
        self.y = y
        self.r = r

        self.xs = [] 
        self.ys = []


        # Variation 1. I prefer variation 1 than 2.
        current_degree = 0  # The degree range to cover is  0º - 360º

        while current_degree <= 360:
            x = self.x + self.r * math.cos(degrees_to_radians(current_degree))
            y = self.y + self.r * math.sin(degrees_to_radians(current_degree))
            self.xs.append(x)
            self.ys.append(y)
            current_degree += dist


        # # Variation 2
        # current_degree = 0
        #
        # while True:
        #     theta = degrees_to_radians(current_degree)
        #     self.xs.append((r * math.cos(theta)) + x)
        #     self.ys.append((r * math.sin(theta)) + y)
        #     current_degree += dist
        #
        #     if current_degree > 360:
        #         break


    def show_graph(self):
        margin = self.r + 1  # The graph area required is about the size of the graph. Thus margin is r+1.
        plt.grid(True)
        axes = plt.gca()

        # Set the size of the graph by adding and subtracting as much as the margin from the center of the circle.
        axes.set_xlim([self.x - margin, self.x + margin])
        axes.set_ylim([self.y - margin, self.y + margin])
        plt.plot(self.xs, self.ys, '.')
        # plt.cla()  # This method initiates the graphs.


# Direction no. 2
CircleEquation(x=0, y=0, r=1, dist=5).show_graph()
````

Draw a circle that has (5, 5) as its center and the radius is 5.

circle_3.png

````python
import math
from matplotlib import pyplot as plt


def degrees_to_radians(degrees):
    return degrees * (math.pi / 180.0)

class CircleEquation:
    def __init__(self, x=0, y=0, r=1, dist=1):  
        self.x = x
        self.y = y
        self.r = r
        
        self.xs = [] 
        self.ys = [] 
        
        deg = 0

        while True:
            theta = degrees_to_radians(deg)
            self.xs.append((r * math.cos(theta)) + x)
            self.ys.append((r * math.sin(theta)) + y)
            deg += dist
            if deg > 360:
                break


    def show_graph(self):
        margin = self.r + 1 
        plt.grid(True)
        axes = plt.gca()
        axes.set_xlim([self.x - margin, self.x + margin]) 
        axes.set_ylim([self.y - margin, self.y + margin])
        plt.plot(self.xs, self.ys, '.')


CircleEquation(x=5, y=5, r=5, dist=1).show_graph()
````
