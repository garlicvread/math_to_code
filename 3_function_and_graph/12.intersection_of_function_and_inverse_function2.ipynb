## The intersection of a function and its inverse function (2)

The intersection of a function and its inverse is equal to the intersection of the original function and `y=x`.

Find the intersection of the function $ g(x) = x^2 + 1 $ $ (x\geq0) $ and its inverse function then return it.

```python
import numpy as np
import matplotlib.pyplot as plt

# g(x)=x^2+1
def g(x):
    return x**2 + 1

# The inverse function g^-1(x)=sqrt(x-1)
# The range of x is the same or greater to zero.
def g_inv(x):
    return np.sqrt(x-1)

# Identity function: y=x
def identity(x):
    return(x)

# g=g^-1
x=np.linspace(0, 6, 601)
graph,= plt.plot(x, g(x))
graph1,= plt.plot(x, g_inv(x))

graph2,= plt.plot(x, identity(x), '--')
plt.legend(handles=(graph,graph1,graph2),  # Add a legend to the graph
           labels=(r'$y=f(x)$', r'$y=f^{-1}(x)$', r'$y=x$'))

# Limit the y-axis range
plt.ylim(0, 6)
plt.show()
```
