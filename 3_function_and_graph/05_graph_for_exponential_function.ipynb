The exponential function is a mathematical function denoted by $$f(x) = exp(x)$$ or $$f(x)=e^x$$ where the argument x is written as an exponent.<br><br>

In general, the exponential function is a transcendental function that uses the exponent of a power as a variable and defines the domain as the whole real number. It is the inverse of the log function.<br><br>

Therefore, an exponential function is a function given as $$y=a^x$$ where a is a positive constant and the domain of x is all real numbers.<br><br>

The equation above is an exponential function about constant a, because when the equation is given, the value of x is set to one value. In this case, the given function is called an exponential function with base a.<br><br>

Here, you will define a class for exponential function which is displayed as
$$y=a^{x+dx}+c$$<br><br>

The arguments a, dx, and c are equal to a, dx, and c of $$y=a^{x+dx}+c$$
which is given above.<br><br>

Graphs of exponential functions can be drawn by while loop.


```python
"""To implement the exponential function, you need to use the 'pow' method from 'Math' module.

Example: 2^3

>>> import math
>>> print(math.pow(2, 3))

    --> The result is 8.
"""


# Variation 1

import math
from matplotlib import pyplot as plt


class ExponentialEquation:
    def __init__(self, a, dx=0, c=0, dist=1, size=100):
        self.a = a
        self.dx = dx
        self.c = c
        self.dist = dist
        self.size = size

        x = -size
        self.xs = []
        self.ys = []

        while x < size:
            x = x + dist
            y = self.a ** (x+self.dx) + self.c
            self.xs.append(x)
            self.ys.append(y)

    def show_graph(self): 
        plt.plot(self.xs, self.ys, '.')
        plt.grid(True)
        axes = plt.gca()
        axes.set_xlim([-self.size, self.size])
        axes.set_ylim([-self.size, self.size])
        plt.show()


# Generate an exponential function that has arguments a = 2 , dx = 0 , c = 0 , dist = 0.1, size = 10
# then display the graph via 'show_graph' method.
ExponentialEquation(2, 0, 0, 0.1, 10).show_graph()
```

```python
# Variation 2

import math
from matplotlib import pyplot as plt

class ExponentialEquation:
    def __init__(self, a, dx=0, c=0, dist=1, size=100):
        self.size = size
        self.xs = []
        self.ys = []

        x = -size

        while True:
            self.xs.append(x)
            self.ys.append(math.pow(a, x+dx) + c)
            x += dist

            if x > size:
                break

    def show_graph(self): 
        plt.plot(self.xs, self.ys, '.')
        plt.grid(True)
        axes = plt.gca()
        axes.set_xlim([-self.size, self.size])
        axes.set_ylim([-self.size, self.size])
        plt.show()


# Generate an exponential function that has arguments a = 3 , dx = 1 , c = 3 , dist = 0.1, size = 10
# then display the graph via 'show_graph' method.
ExponentialEquation(3, 1, 3, 0.1, 10).show_graph()
```
