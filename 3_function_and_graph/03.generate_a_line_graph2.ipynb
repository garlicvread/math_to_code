## Drawing a line graph (2)
Find the domain and range of the following function then draw a line graph for it.

$$ g(x) = \sqrt{x−1} $$

To define the function ***g***, ***x-1*** must be the same or greater than 0,
and the range of the value of the function ***g(x)*** is the same or greater than 0 as well.

Thus, the domain of the function g is: $$ \lbrace x|x≥1 \rbrace $$
and the range of the function g is: $$ \lbrace y|y≥0 \rbrace $$


## Instruction
1. Use **`np.linspace()`** method to define the range x-axis of the function ***g*** and store it in variable ***x***.
3. Use **`np.sqrt()`** method to define the function ***g*** for the variable ***x***.

```python
import numpy as np
import matplotlib.pyplot as plt

# Using np.linspace() to define the range of x-axis and store it in x.
x = np.linspace(1, 10, 901)

# Using np.sqrt() to calculate the square root of x-1 and store it in gx.
gx = np.sqrt(x-1)

plt.plot(x, gx)
plt.show()
```
