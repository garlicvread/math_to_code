### Drawing a line graph (1)
For the given function ***f(x) = 3x+1***, the function ***f*** is defined by all real numbers.

The image of ***x*** by the function ***f*** is ***3x+1***, and it is also defined by all real numbers.

Thus, the domain and range of the given function are the set of all real numbers ***R***.

<br>

### Instruction
1. Use **`np.linspance()`** method to define the range of x-axis and store it in variable ***x***.

2. Define the function ***f*** in the variable ***fx***.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the range of x as from negative infinity to positive infinity.
x = np.linspace(-5, 5, 1001)

# Define a function of f(x) = 3x + 1
fx = 3 * x + 1

plt.plot(x, fx)
plt.grid(True)

plt.show()
```
